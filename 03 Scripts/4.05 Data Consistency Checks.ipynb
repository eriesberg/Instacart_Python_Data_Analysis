## Question 3.
Check for mixed-type data in your df_ords dataframe.

````python
def mixed_types(x):
    for col in x.columns.tolist():
        weird = (x[[col]].applymap(type) != x[[col]].iloc[0].apply(type)).any(axis = 1)
        if len (x[weird]) > 0:
            print(col)    
````

````python
mixed_types(df_ords)
````
